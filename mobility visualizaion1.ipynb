```python
# Re-importing the pandas library
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Re-creating the DataFrame from the provided data
data = "drawing.csv"
df = pd.read_csv(data)

# Plotting the data
plt.figure(figsize=(18, 8))
plt.rcParams.update({'font.size': 16})
# Plotting both 'average_pop_during' and 'average_pop_before'
bar_width = 0.4
index = np.arange(len(df['zipcode']))
bar1 = plt.bar(index, df['average_pop_before'], bar_width, label='Average Pop Before Covid', color='blue', alpha=0.6)
bar2 = plt.bar(index + bar_width, df['average_pop_during'], bar_width, label='Average Pop During Covid', color='orange', alpha=0.6)


plt.xlabel('Zipcode')
plt.ylabel('Average Population')
plt.title('Comparison of Average Population Before and During Covid 1')
plt.xticks(index + bar_width / 2, df['zipcode'], rotation=45)
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Plotting only the line plots for a clearer trend visualization

plt.figure(figsize=(18, 8))
plt.rcParams.update({'font.size': 16})

# Create lines
line1 = plt.plot(index, df['average_pop_before'], color='blue', marker='x', label='Average Pop Before Covid', linewidth=2)
line2 = plt.plot(index, df['average_pop_during'], color='orange', marker='o', label='Average Pop During Covid', linewidth=2)

plt.xlabel('Zipcode')
plt.ylabel('Average Population')
plt.title('Comparison of Average Population Before and During Covid 2')
plt.xticks(index, df['zipcode'], rotation=45)
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```
