```python
import pandas as pd
import numpy as np
from google.colab import drive

drive.mount('/content/drive')
```

```python
file_path = '/content/drive/MyDrive/CIEN 4012 Project/Data/Citibike raw data/201906.csv'
bike_data = pd.read_csv(file_path)

bike_data
```

```python
def haversine(lat1, lon1, lat2, lon2):
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2.0)**2
    c = 2 * np.arcsin(np.sqrt(a))
    # Earth radius 6378km
    km = 6378 * c
    return km

bike_data['trip_distance_km'] = haversine(
    bike_data['start_station_latitude'],
    bike_data['start_station_longitude'],
    bike_data['end_station_latitude'],
    bike_data['end_station_longitude']
)

bike_data.head()
```

```python
# Group the data by 'zipcode' and calculate the average distance for each group
average_distance_per_zipcode = bike_data.groupby('zipcode')['trip_distance_km'].mean().reset_index()

average_distance_per_zipcode
```

```python
# Count the number of unique start and end stations per zipcode by creating a set of station IDs for each
bike_data['start_station'] = bike_data['start_station_id'].astype(str) + ' - ' + bike_data['start_station_name']
bike_data['end_station'] = bike_data['end_station_id'].astype(str) + ' - ' + bike_data['end_station_name']

# Combine the start and end station sets for each zipcode and count unique stations
station_count_per_zipcode = bike_data.groupby('zipcode').apply(
    lambda x: len(set(x['start_station']).union(set(x['end_station'])))
).reset_index(name='number_of_stations')

station_count_per_zipcode
```
