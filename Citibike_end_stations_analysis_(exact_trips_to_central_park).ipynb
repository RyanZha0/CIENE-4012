```python
from shapely.geometry import Point, Polygon

# Define Manhattan's geographic boundaries with a simplified polygon approximation.
manhattan_polygon_points = [
    (-73.934580, 40.880912),
    (-73.951977, 40.851736),
    (-74.013391, 40.754845),
    (-74.024283, 40.698188),
    (-74.009793, 40.696703),
    (-73.996477, 40.705462),
    (-73.973175, 40.708431),
    (-73.965734, 40.730099),
    (-73.965358, 40.743476),
    (-73.939437, 40.775202),
    (-73.924349, 40.798349),
    (-73.933169, 40.809234),
    (-73.933984, 40.833898),
    (-73.907016, 40.874670)
]
manhattan_polygon = Polygon(manhattan_polygon_points)

central_park_polygon_points = [
    (-73.9580, 40.8007),
    (-73.9493, 40.7968),
    (-73.9720, 40.7643),
    (-73.9821, 40.7676)
]
central_park_polygon = Polygon(central_park_polygon_points)

# Function to determine if a given point is within the specified polygon
def is_in_polygon(lng, lat, polygon):
    point = Point(lng, lat)
    return polygon.contains(point)

# Load the dataset
file_path = '201906_01.csv'  # Adjust the file path accordingly
df = pd.read_csv(file_path)

# Filter for records where the origin is within Manhattan and the destination is within Central Park
filtered_data = df[
    df.apply(lambda x: is_in_polygon(x['start_station_longitude'], x['start_station_latitude'], manhattan_polygon) and
             is_in_polygon(x['end_station_longitude'], x['end_station_latitude'], central_park_polygon), axis=1)
]
filtered_data.to_csv('201906_centralpark.csv', index=False)
filtered_data
```

```python
import shapely.wkt
import numpy as np

data = filtered_data.sample(frac=0.1, random_state=42)

reduced_file_path = 'reduced_201906.csv'
data.to_csv(reduced_file_path, index=False)

reduced_file_path, data.shape
```

```python
# Load datasets
file_path = 'reduced_201906.csv'
zip_code_file_path = '/content/drive/MyDrive/CIEN 4012 Project/Codes/mobility data - during covid/Modified_Zip_Code_Tabulation_Areas__MODZCTA_.csv'

data = pd.read_csv(file_path)
zip_codes_data = pd.read_csv(zip_code_file_path)

# Convert the_geom column to Shapely Polygon objects in zip_codes_data
zip_codes_data['polygon'] = zip_codes_data['the_geom'].apply(lambda wkt_str: shapely.wkt.loads(wkt_str))

# Initialize a column for zip codes in the original data
data['zipcode'] = np.nan

# Match zip codes to the starting points
for idx, row in data.iterrows():
    point = Point(row['start_station_longitude'], row['start_station_latitude'])
    for _, zip_row in zip_codes_data.iterrows():
        polygon = zip_row['polygon']
        if polygon.contains(point):
            data.at[idx, 'zipcode'] = zip_row['MODZCTA']
            break
# Drop rows with NaN zip codes and finalize
data.dropna(subset=['zipcode'], inplace=True)
data['zipcode'] = data['zipcode'].astype(int)

data.to_csv('201906_final.csv', index=False)
data
```

```python
import folium
from folium import plugins

# Create a map centered around an average location in the data
avg_lat = (data['start_station_latitude'].mean() + data['end_station_latitude'].mean()) / 2
avg_lon = (data['start_station_longitude'].mean() + data['end_station_longitude'].mean()) / 2
map_trip = folium.Map(location=[avg_lat, avg_lon], zoom_start=13)

# Drawing lines for each trip from start to end station
for _, row in data.iterrows():
    start_coords = [row['start_station_latitude'], row['start_station_longitude']]
    end_coords = [row['end_station_latitude'], row['end_station_longitude']]
    line = folium.PolyLine(locations=[start_coords, end_coords], weight=0.4, color='blue')
    map_trip.add_child(line)

map_trip
```

```python
# Extract unique end stations with their names and coordinates
unique_end_stations = data[['end_station_id', 'end_station_name', 'end_station_latitude', 'end_station_longitude']].drop_duplicates()

# Count the number of unique end stations
num_end_stations = unique_end_stations.shape[0]

unique_end_stations, num_end_stations
```

```python
from IPython.display import display, HTML
# Count the number of trips that end at each station
end_station_counts = data['end_station_id'].value_counts().reset_index()
end_station_counts.columns = ['end_station_id', 'number_of_trips']

# Merge this count with the unique end stations dataframe
end_station_summary = pd.merge(unique_end_stations, end_station_counts, on='end_station_id')

# Convert the merged dataframe to an HTML table
html_table_with_counts = end_station_summary.to_html(index=False, classes='table table-striped', border=0)

# Display the HTML table
display(HTML(html_table_with_counts))
```

```python
import matplotlib.pyplot as plt

# Sort the dataframe by 'number of trips' in descending order
end_station_summary_sorted = end_station_summary.sort_values('number_of_trips', ascending=False)
end_station_summary_sorted

```

```python
end_station_summary_sorted.to_csv('201906_sorted.csv', index=False)
```

```python
# Plotting the data
plt.figure(figsize=(10, 8))
plt.bar(end_station_summary_sorted['end_station_name'], end_station_summary_sorted['number_of_trips'][:20], color='skyblue')
plt.xlabel('End Station Name', fontsize=14)
plt.ylabel('Number of Trips', fontsize=14)
plt.title('End Stations by Number of Trips', fontsize=16)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import folium

# Create a new base map centered around the general location of the stations
m_points = folium.Map(location=[40.73, -73.93], zoom_start=12)

# Add a marker for each station with size and color proportional to the number of trips
for index, row in end_station_summary_sorted.iterrows():
    # Scaling the number of trips to get a reasonable size for visualization
    radius = row['number_of_trips'] / 8  # Scale factor for size
    color_intensity = min(row['number_of_trips'] / 142 * 255, 255)  # Max trips in data is 142, scale color accordingly
    # Convert numeric intensity to hex color
    hex_color = f'#{int(color_intensity):02x}00{255-int(color_intensity):02x}'

    folium.CircleMarker(
        location=(row['end_station_latitude'], row['end_station_longitude']),
        radius=radius,  # Dynamic radius based on number of trips
        color=hex_color,  # Dynamic color based on number of trips
        fill=True,
        fill_color=hex_color
    ).add_to(m_points)

# Display the map
m_points
```
